```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
plt.xlabel('x축', color='red', loc='right') # loc='left','center','right'
plt.ylabel('y축', color='#00aa00', loc='top') #loc='bottom','center','top'
```

## 막대 그래프

```python
label = ['강백호','서태웅','정대만']
values = [190, 187, 184]
plt.bar(label, values)
```

```python
label = ['강백호','서태웅','정대만']
values = [190, 187, 184]
colors = ['r','g','b']
plt.bar(label, values, color=colors, alpha=0.5, width=0.5)
plt.ylim(175, 195)
```

```python
label = ['강백호','서태웅','정대만']
values = [190, 187, 184]

bar = plt.barh(label, values)

bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[1].set_hatch('.')

plt.xlim(175, 195)
```

```python
label = ['강백호','서태웅','정대만']
values = [190, 187, 184]

bar = plt.bar(label, values)

plt.ylim(175, 195)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha='center',color='black')
```

## DataFrame에 활용

```python
import pandas as pd

df = pd.read_csv('nado_data.csv')

plt.bar(df['이름'], df['국어'])
# plt.bar(df['이름'], df['영어'], bottom=df['국어'])
plt.bar(df['이름'], df['영어'], bottom=df['국어'] + df['영어'])
```

```python
plt.bar(df['이름'], df['국어'], label='국어')
plt.bar(df['이름'], df['영어'], bottom=df['국어'], label='영어')
plt.bar(df['이름'], df['영어'], bottom=df['국어'] + df['영어'], label='수학')

plt.xticks(rotation=30)
plt.legend()
```

## 다중 막대 그래프

```python
import numpy as np # type: ignore

print(np.arange(5))
print(np.arange(3, 6))

arr = np.arange(6)
(arr + 100) * 3
```

```python
index = np.arange(df.shape[0])

w = 0.25

plt.bar(index - w, df['국어'], label='국어', width=0.25)
plt.bar(index, df['영어'], label='영어', width=0.25)
plt.bar(index + w, df['수학'], label='수학', width=0.25)
plt.ylim(0, 100)
plt.xticks(rotation=30)
plt.legend()
```
